```python
from yolov5 import YOLOv5
import cv2 
from yolov5.utils.plots import colors, plot_one_box
from yolov5.utils.general import xywh2xyxy
import numpy as np
import torch
```

```python
model_path = "~/Developer/yolov5/yolov5s.pt"
device = 'cpu'
cam = cv2.VideoCapture(0)
net = YOLOv5(model_path, device)
```

```python
def get_detections(net, img):
    gn = torch.tensor(img.shape)[[1, 0, 1, 0]]
    detections = net.predict(img)
    for bbox in detections.xywh:
        for *xywh, cond, cls in bbox:
            if cls == 0 and cond > 0.7:
                x,y,w,h = int(xywh[0]),int(xywh[1]),int(xywh[2]),int(xywh[3])
                xyxy = xywh2xyxy(torch.tensor(xywh).unsqueeze(0))
                print(xyxy)
                label = f'{detections.names[int(cls)]} {cond:.2f}'
                plot_one_box(xyxy[0], img, label=label, color=colors(int(cls),True), line_thickness=3)
    return img
```

```python
while True:
    ret, frame = cam.read()
    if not ret:
        break
    img = get_detections(net, frame)
    cv2.imshow("test", img)

    k = cv2.waitKey(1)
    if k%256 == 27:
        print("Escape hit, closing...")
        break
cam.release()

cv2.destroyAllWindows()
```
